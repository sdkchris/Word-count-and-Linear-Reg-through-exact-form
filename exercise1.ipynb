**Word Count Program:** 
In this task your are required to use the provided text file and write a program that will count the number of occurances of unique words.

The program should ignore words like {’the’, ’a’,’an’,’be’}
Finally you are required to generate the histogram of the top 10 most occuring words. 
Click here to download Text File(https://raw.githubusercontent.com/tensorflow/models/master/research/object_detection/README.md)



Let's first prepocess the test file by: 

  1. Removing special character such as {^°!§$%&/()=?\@€µ+*.,;:_[]}
  2. Removing stop words such as ['the','a','an', 'be','img','br','jpg','b', 'p']
  3. Setting the characters to lowecase

```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```
file = "/content/drive/My Drive/README.txt"

#list containing stop words to be removed from the file
stopWords = ['the','a','an', 'be','img','br','jpg','b','p', '2','we']

#list containing special characters to be removed
specialChars = "{^°!§$%&/()=?\@€µ+*.,;:_[]#<>|}"
```

```
#read file
txtfile = open(file, 'r').read()

#1. let's remove special characters in the file and replace them with ""
for s in specialChars:
  txtfile = txtfile.replace(s,' ')

#2. let's set all characters to lowercase
txtfile = txtfile.lower()

#3. split text file into a list of words
listOfWords = txtfile.split()

#4. Now we remove stopwords from the txt file
for word in listOfWords:
  for sW in stopWords:
    if word == sW:
      listOfWords.remove(word)
```

```
print(listOfWords)
```

```
#Count number of occurences of unique words
#we'll use a dictionary--> key-value pairs where the keys are words and values are the number of times these words occur

WordDict = {}

for word in listOfWords:
  WordDict[word] = WordDict.get(word,0) + 1

#view dict
#WordDict

#let's sort the list and get top 10 most occuring words
WordDict = sorted(WordDict.items(), key=lambda x: x[1] ,reverse=True)
#top 10 words
top10 = WordDict[0:10]
print("10 Most occuring words")
top10
```

```
#let's plot their occurence on a histogram 
wordList = []
wordCount = []

for i in top10:
  wordList.append(i[0])
  wordCount.append(i[1])
#plot size parameters
fig_size = plt.rcParams['figure.figsize']
fig_size[0] = 8
fig_size[1] = 5
plt.rcParams['figure.figsize']=fig_size
plt.bar(wordList, wordCount,1,color='r')
plt.show()
```

**Matrix Multiplication:** 
You are required to use numpy for operation on matrices. Create a matrix A of dimensions n × m, where n = 100 and m = 20. Initialize Matrix A. Create a vector v of dimension m × 1. Initialize the matrix with a random values and vector with normal distribution using μ = 2 and σ = 0.01. 

Perform following operation on them

– Iterative multiply (element-wise) each row of matrix A with vector v and sum the result of each iteration in another vector c
– Find mean and standard deviation of the new vector c 
– Plot histogram of vector c using 5 bins

```
#matrix A
m = 20
n = 100
mu = 2
sigma = 0.01
matA = np.random.random((n,m)) 
#vector v
vec_v = np.random.normal(mu,sigma,m )

#element-wise multiplication
c = []
for i in range(n): #row
  temp = 0
  for j in range(m): #col
    temp = temp + matA[i][j] * vec_v[j]
  c.append(temp)
print(c)
```

```
#mean  of C
mean_c = np.mean(c)
print("Mean of c is {}".format(mean_c))

#standard deviation of c
standdv_c = np.std(c)
print("Standard deviation of c is {}".format(standdv_c))
```

```
#plot histogram of vec c using 5 bins
plt.hist(c, bins=5, color='r')
plt.show()
```

#2 Linear Regression through Exact Form

In this exercise you will implement linear regression that was introduced in the introduction Machine Learning Lecture.

Generate 3 sets of simple data. i.e. a matrix A with dimensions 100 × 2. Initialize it with normal distribution μ = 2 and σ = [0.01, 0.1, 1]

```
c,d = 100,2
mu = 2
std1,std2,std3 = 0.01, 0.1 ,1
mat_A1 = np.random.normal(mu, std1,(c,d))
mat_A2 = np.random.normal(mu, std2,(c,d))
mat_A3 = np.random.normal(mu,std3,(c,d))
```

```
mat_A1.shape
```

Implement LEARN-SIMPLE-LINREG algorithm and train it using matrix A to learn values of β0 and β1

simple_linear_reg_formula.gif

```
#function to learn simple linear regression 

def learn_simple_linreg(matrix):
  #get features and target
  x = matrix[:,0]
  y = matrix[:,1]
  #get mean of x and y
  x_mean = np.mean(x)
  y_mean = np.mean(y)

  #calculate the slope of the linear line or b1
  b1 = sum((x-x_mean)*(y-y_mean))/sum((x-x_mean)**2)
  #compute intercept or b0
  b0 = y_mean- (b1 * x_mean)
  return b0, b1
```

Implement PREDICT-SIMPLE-LINREG and calculate the points for each training example in matrix A.

```
def predict_simple_linreg(matrix):
  x = matrix[:,0]
  y = matrix[:,1]
  b0, b1 = learn_simple_linreg(matrix)
  
  #computing y_pred
  y_pred = b0 + b1*x
  return y_pred
```

```
#prdict values of all three matrices
y_pred_A1 = predict_simple_linreg(mat_A1)
y_pred_A2 = predict_simple_linreg(mat_A2)
y_pred_A3 = predict_simple_linreg(mat_A3)
```

Plot the training points from matrix A and predicted values in the form of line graph.

```
#figure parameters 
fig_size = plt.rcParams['figure.figsize']
fig_size[0]=6
fig_size[1]=10
plt.rcParams['figure.figsize'] = fig_size

#graph specific parameters
f, axarr = plt.subplots(3,sharex=False)
axarr[0].set_title("Model Comparison")
axarr[0].plot(range(100), mat_A1[:,1], color = 'red')
axarr[0].plot(range(100), y_pred_A1, color = 'blue')
axarr[0].set(ylabel = "matrix A, sigma = 0.01")

axarr[1].plot(range(100), mat_A2[:,1], color = 'red')
axarr[1].plot(range(100), y_pred_A2, color = 'blue')
axarr[1].set(ylabel = "matrix A, sigma = 0.1")

axarr[2].plot(range(100), mat_A3[:,1], color = 'red')
axarr[2].plot(range(100), y_pred_A3, color = 'blue')
axarr[2].set(ylabel = "matrix A, sigma = 1")

plt.show()
```

Notice that as the value of sigma increases, the variance of the predicted values decreases. Also, predicted values of Matrix A with sigma = 0.01, tends to be a straight horizontal line as compared to the Matrix A with sigma = 1.

In the end use numpy.linalg lstsq to replace step 2 for learning values of β0 and β1

```
#let's now use numpy.linalg to learn b0 and b1
def np_linalg_lstsq(matrix):
  x = matrix[:,0]
  y = matrix[:,1]

  A = np.vstack([x, np.ones(matrix.shape[0])]).T
  b1, b0 = np.linalg.lstsq(A,y)[0] #returns a least quare solution to a linear matrix solution 
  return b1,b0
```

```
#prediction function using np.linalg.lstsq
def predict_np_linalg(matrix):
  x = matrix[:,0]
  b1, b0 = np_linalg_lstsq(matrix)
  y_pred = b0 + b1*x
  return y_pred
```

```
#prdict values of all three matrices with np.linalg.lstsq
y_pred_A1 = predict_np_linalg(mat_A1)
y_pred_A2 = predict_np_linalg(mat_A2)
y_pred_A3 = predict_np_linalg(mat_A3)
```
